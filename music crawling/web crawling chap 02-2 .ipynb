```python
from selenium import webdriver
driver = webdriver.Chrome('C:/chromedriver.exe')
```

```python
# 웹 페이지 접속98
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%BD%94%EB%A1%9C%EB%82%98'
driver.get(url)
```

```python
# 웹 페이지 다운로드 
nn = driver.page_source
    
from bs4 import BeautifulSoup
soup = BeautifulSoup(nn, 'html.parser')
# get_text() = > 텍스트 가져옴 

al = soup.find_all('ul')# 맨앞 하나의 값만 추출
for i in al:
    print(i.get_text()) 
```

```python
# 정보 가져오기 - 태그 그룹에서 하나의 태그 선택하기 
tags = soup.select('.info_01')
tag_1 = tags[0]

tag_1
```
