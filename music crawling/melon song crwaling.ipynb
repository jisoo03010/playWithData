```python
# 멜론 인기 노래 순위로 크롤링 해보기
from selenium import webdriver 

driver = webdriver.Chrome('C:/Users/sdh20/Desktop/playwithdata/chromedriver.exe')
url  = 'https://www.melon.com/chart/index.htm'

driver.get(url)

from bs4 import BeautifulSoup
html = driver.page_source
soup= BeautifulSoup(html, 'html.parser')

song = soup.select('tbody > tr')
# print(len(song)) # 100

song_data =[]
rank =1
for i in song:
    title = i.select('div.ellipsis.rank01 > span >a')[0].text
    sing = i.select('div.ellipsis.rank02 >a')[0].text
    rank = rank + 1
    song_data.append(['melon', rank, title, sing])
    
import pandas as pd
columns = ['서비스', '순위', '타이틀', '가수']
pd_data = pd.DataFrame(song_data, columns = columns)    
pd_data
```

```python
pd_data.to_excel('melon.xlsx', index = True)
```
