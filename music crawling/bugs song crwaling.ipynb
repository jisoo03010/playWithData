```python
from selenium import webdriver
driver = webdriver.Chrome('C:/Users/sdh20/Desktop/playwithdata/chromedriver.exe')
url = 'https://music.bugs.co.kr/chart'
driver.get(url)
from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 100개의 인기차트 값 가져오기
```

```python
songs = soup.select('table.byChart > tbody > tr')


song_data =[]
rank =0
for i in songs:
    title = i.select('th > p.title > a')[0].text
    sing = i.select('td > p.artist > a')[0].text
    rank = rank + 1
    song_data.append(['bugs',rank, title, sing])
    
columns = ['서비스', '순위', '타이틀', '가수']
import pandas as pd

bugsDataFrame = pd.DataFrame(song_data, columns = columns)
```

```python
bugsDataFrame.to_excel('bugs.xlsx', index = True)
```
