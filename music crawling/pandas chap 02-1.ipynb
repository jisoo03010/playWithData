```python
# 국적코드와 입국객수 둘다 보고싶은 경우 ? => 대괄호를 2번 묶어서 실행
sample_1[['입국객수', '국적코드']]
```

```python
# 컬럼 생성 
sample_1['기준년월'] = '2019-11'
sample_1
```

```python
condition = (sample_1['성별'] == '남성')
sample_1[condition]
```

```python
condition = (sample_1['입국객수'] >= 150000)
sample_1[condition]
```

```python
# 성별이 남성이면서, 입국 객수가 150000이상인 경우 
condition = (sample_1['성별'] =='남성' ) & (sample_1['입국객수'] >= 150000)
sample_1[condition]
```

```python
# 두 개의 컬럼에 대해 필터링하기 - 가독성 높게 표현하기
#둘다 만족 : &
# 역슬래시 기호 : 코드가 길어서 한 줄 아래에 이어서 작성할 테니 두 줄을 한 줄로 인식하라는 의미 
condition = (sample_1['성별'] =='남성' ) \
    & (sample_1['입국객수'] >= 150000)
sample_1[condition]
```

```python
# 한 가지만 만족 : |
condition = (sample_1['성별'] =='남성' ) \
    | (sample_1['입국객수'] >= 150000)
sample_1[condition]
```

```python
condition  = (sample_1['국적코드'].isin(['A01', 'A18']))
sample_1[condition]
```

```python
sample_1
```

```python
# A01 , A18에 해당하지 않는 경우만 보고 싶은 경우에는 
sample_1[condition == False]
```
